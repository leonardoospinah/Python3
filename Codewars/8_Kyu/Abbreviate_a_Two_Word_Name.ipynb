```python
# My solution
def abbrev_name(name):
    new_name = ""
    for x in name.upper().split():
        new_name += x[0]+"."
    return new_name[:-1]
```

```python
# Community solution
def abbrevName(name):
    return '.'.join(w[0] for w in name.split()).upper()
```
