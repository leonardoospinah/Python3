```python
# My solution
def uni_total(s):
    total = 0
    if s == "":
        return 0
    elif len(s) > 1:
        for x in s:
            total += ord(x)
        return total
    else:
        return ord(s)
```

```python
# Community solution
def uni_total(string):
    return sum(map(ord, string))
```
